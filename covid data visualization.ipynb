```python
# importing important libraries

import pymysql
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# establishing connection with database server
conn = pymysql.connect(host='localhost',user='root',passwd='samir@MySQL',database='state_daily_update')
```

```python
# defining cursor to run command
cursor = conn.cursor()
cursor.execute("select TT, Date_YMD, Status from state_wise_daily")
```

```python
# fetching the data from server using cursor
result = cursor.fetchall()
```

```python
# defining variable & array to store number of confirmed cases, number of recovered, and number of deceased cases per day
daily_confirmed = []
daily_recovered = []
daily_deceased = []
date = []        
total_confirmed = 0
total_recovered = 0
total_deceased = 0
for i in result:
    if i[2] == "Confirmed":
        daily_confirmed.append(i[0])
        date.append(i[1])
        total_confirmed += int(i[0])
    elif i[2] == "Recovered":
        daily_recovered.append(i[0])
        total_recovered += int(i[0])
    elif i[2] == "Deceased":
        daily_deceased.append(i[0])
        total_deceased += int(i[0])
```

```python
# plt.bar to plot a bar graph
# with given values

# try to display date at larger interval
index = pd.date_range(start=date[0], end=date[-1], freq="D")
df = pd.DataFrame(data=daily_confirmed, index=index, columns=['daily_confirmed'])
ax = df.plot(color='purple')
plt.xticks(rotation=90) 
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
date_form = mdates.DateFormatter("%d-%m-%Y")
ax.xaxis.set_major_formatter(date_form)

# setting xlabel of graph
plt.xlabel("Date")

# setting ylabel of graph
plt.ylabel("Daily confirmed cases")

# setting title of graph
plt.title("Confirmed Cases vs Time", fontsize = "20")

# show() method to display the graph
plt.show()
```
